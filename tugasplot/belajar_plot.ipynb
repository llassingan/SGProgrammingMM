# Menggambar Plot


```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Menyiapkan data (sin(2wt + theta))
def sinus(amp,frek,t,theta):
	t = np.arange(0,t,0.1)
	y = amp * np.sin(2*frek*t + np.deg2rad(theta))
	return t,y
```

```
# plotting
t1,y1 = sinus(1,1,4,0)
plt.plot(t1,y1)

t2,y2 = sinus(1,1,4,30)
plt.plot(t2,y2, 'r')

t3,y3 = sinus(1,1,4,60)
plt.plot(t3,y3, 'b--')

t4,y4 = sinus(1,1,4,90)
plt.plot(t4,y4, 'c-o')
plt.show()
```
